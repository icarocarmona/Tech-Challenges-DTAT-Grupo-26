Utilizamos um pipeline composto por funes para construir, treinar, gerar previses e avaliar o desempenho de um modelo LSTM na previso da bolsa de valores. A escolha da arquitetura LSTM, uma variante das Redes Neurais Recorrentes (RNNs),  baseada em caractersticas que se mostraram eficazes para lidar com dados temporais e sequenciais. Alguns motivos para essa escolha incluem:

Manejo de Sequencialidade: RNNs, especialmente LSTMs, so adequadas para lidar com a natureza temporal das sries temporais da bolsa de valores, capturando padres e dependncias ao longo do tempo.

Memria de Longo Prazo: LSTMs superam problemas de dependncias temporais de longo prazo, sendo capazes de manter e atualizar a memria ao longo do tempo.

Tratamento de Dados Irregulares: LSTMs podem aprender padres complexos e se adaptar a mudanas repentinas, tornando-as robustas para lidar com a volatilidade e irregularidades dos dados da bolsa de valores.

Capacidade de Modelar Relaes No-Lineares: LSTMs, com suas unidades de memria e portes, so capazes de modelar relaes no-lineares complexas nos dados da bolsa de valores.

Aprendizado Contnuo: RNNs, incluindo LSTMs, suportam o aprendizado contnuo, permitindo a adaptao do modelo a novas informaes sem a necessidade de treinamento completo.

 crucial observar que, embora RNNs e LSTMs tenham se mostrado eficazes em tarefas temporais, a previso da bolsa de valores  desafiadora devido  sua natureza incerta e influenciada por diversos fatores externos. O desempenho do modelo depende da qualidade e quantidade dos dados, sendo essencial realizar anlises cuidadosas e considerar estratgias adicionais, como pr-processamento adequado e otimizao de hiperparmetros.

As Bandas de Bollinger consistem em trs linhas: a SMA (mdia mvel simples), uma banda superior e uma banda inferior. As bandas superior e inferior so calculadas em relao  SMA e ao desvio padro dos preos de fechamento. Vamos  calcular as Bandas de Bollinger com um desvio padro de 2 em relao  SMA de 20 dias (inputs padres utilizado na plataforma https://br.investing.com/indices/bovespa-chart):

A ideia  adicionar essa nova feture para melhorar o modelo


```python
n = 20  # Perodo da SMA
k = 2  # Nmero de desvios padro

data['SMA'] = data['Adj Close'].rolling(window=n).mean()
data['Std'] = data['Adj Close'].rolling(window=n).std()

# Bandas de Bollinger
data['Upper_Bollinger'] = data['SMA'] + (k * data['Std'])
data['Lower_Bollinger'] = data['SMA'] - (k * data['Std'])
del data['Std']
```

```python
from tensorflow.keras.layers import LSTM, Dense, Reshape

# Supondo que `data`  o seu DataFrame e 'lookback_period' e 'forecast_period' so definidos
lookback_period = 60
forecast_period = 3
n_units = 50
batch_size = 64
n_epochs = 200
n_fetures = len(fetures)

adj_close = []
if len(fetures) == 1:
  adj_close = data[fetures].values.reshape(-1, 1)
else:
  adj_close = data[fetures].values

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_close = scaler.fit_transform(adj_close)

X = []
Y = []

y = scaled_close

for i in range(lookback_period, len(y) - forecast_period + 1):
    X.append(y[i - lookback_period:i])
    Y.append(y[i:i + forecast_period])

X_train = np.array(X)
Y_train = np.array(Y)

# Diviso em conjuntos de treino e teste
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]

X_train = np.array(X_train)
Y_train = np.array(Y_train)
X_test = np.array(X_test)
Y_test = np.array(Y_test)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_test shape: {Y_test.shape}")
```

```python
# Modelo
model = Sequential()
model.add(LSTM(units=n_units, return_sequences=True, input_shape=(lookback_period, n_fetures)))
model.add(LSTM(units=n_units))
model.add(Dense(forecast_period * n_fetures)) # Ajustando a camada Dense
model.add(Reshape((forecast_period, n_fetures))) # Adicionando uma camada de redimensionamento
model.summary()
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, Y_train, epochs=n_epochs, batch_size=batch_size, verbose=0)
```

```python
# Fazendo previses
Y_pred = model.predict(X_test)

# Redimensionando Y_test e Y_pred para o clculo do MSE
Y_test_reshaped = Y_test.reshape(Y_test.shape[0], Y_test.shape[1] * n_fetures)
Y_pred_reshaped = Y_pred.reshape(Y_pred.shape[0], Y_pred.shape[1] * n_fetures)

# Calculando o MSE
mse = mean_squared_error(Y_test_reshaped, Y_pred_reshaped)
print(f"MSE: {mse}")
```
